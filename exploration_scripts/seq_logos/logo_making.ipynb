# Three logos in a single image

```python
input_dirpath = '/mnt/1.5_drive_0/RiboGrove_workdirs/4.210/antiSD/draw_seq_logos'
axis_label_fontsize = 14


# == Flavobacteriales ==

f, (ax1, ax2, ax3) = plt.subplots(3, 1, sharex=True, figsize=[14, 10])
f.tight_layout(h_pad=4.5)


my_df, coverages = create_matrix.create_matrix(
    os.path.join(
        input_dirpath,
        'Flavobacteriales_tails21.aln_amended.fasta'
    )
)

my_x_ticks =  np.array([  0,  1,  6, 11, 16, 20])
ecoli_xticks = -np.array([21, 20, 15, 10,  5,  1])

# create a Logo object
my_logo = logomaker.Logo(my_df,
                         ax=ax1,
                         font_name='DejaVu Sans',
                         color_scheme='colorblind_safe',
                         width=.8
)

# style using Logo methods
for i in range(13, 19):
    my_logo.highlight_position(p=i, color='#BFBFBF', alpha=.7)
# end for

# style using Axes methods

my_logo.ax.set_ylabel('Frequency', fontsize=axis_label_fontsize)

my_logo.ax.set_xlabel('Position', fontsize=axis_label_fontsize)
my_logo.ax.set_xticks(
    my_x_ticks
)
my_logo.ax.set_xticklabels(ecoli_xticks)
my_logo.ax.tick_params(labelbottom=True)


# == Sulfolobaceae ==

my_df, coverages = create_matrix.create_matrix(
    os.path.join(
        input_dirpath,
        'Sulfolobaceae_tails21.aln_amended.fasta'
    )
)

my_x_ticks =   np.array([  0,  1,  6, 11, 16, 19, 20])
ecoli_xticks = -np.array([21, 20, 15, 10,  5,  2,  1])

# create a Logo object
my_logo = logomaker.Logo(my_df,
                         ax=ax2,
                         font_name='DejaVu Sans',
                         color_scheme='colorblind_safe',
                         width=.8)

# style using Logo methods
for i in range(13, 19):
    my_logo.highlight_position(p=i, color='#BFBFBF', alpha=.7)
# end for

# style using Axes methods

my_logo.ax.set_ylabel('Frequency', fontsize=axis_label_fontsize)

my_logo.ax.set_xlabel('Position', fontsize=axis_label_fontsize)
my_logo.ax.set_xticks(
    my_x_ticks
)
my_logo.ax.set_xticklabels(ecoli_xticks)
my_logo.ax.set_xlim([-0.4, 20.4])
my_logo.ax.tick_params(labelbottom=True)


# == Carsonella ==

my_df, coverages = create_matrix.create_matrix(
    os.path.join(
        input_dirpath,
        'Carsonella_21tails.aln.fasta'
    )
)

my_x_ticks = np.array([0, 1, 6, 11, 16, 20])
ecoli_xticks = -np.array([21, 20, 15, 10, 5, 1])

# create Logo object
my_logo = logomaker.Logo(my_df,
                         ax=ax3,
                         font_name='DejaVu Sans',
                         color_scheme='colorblind_safe',
                         width=.8)

# style using Axes methods
my_logo.ax.set_ylabel('Frequency', fontsize=axis_label_fontsize)

my_logo.ax.set_xlabel('Position', fontsize=axis_label_fontsize)
my_logo.ax.set_xticks(
    my_x_ticks
)
my_logo.ax.set_xticklabels(ecoli_xticks)
my_logo.ax.tick_params(labelbottom=True)
```
